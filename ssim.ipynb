# Geolife SSIM

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

Dir = './'
df_entropy = pd.read_csv('geo_result.csv')
df_es = df_entropy

import matplotlib.patches as mpatches

df_100 = df_es['100m'].dropna()
df_300 = df_es['300m'].dropna()
df_500 = df_es['500m'].dropna()
df_700 = df_es['700m'].dropna()
df_900 = df_es['900m'].dropna()

# plot
FS = 30
fig, ax = plt.subplots(figsize=(12, 6))
patterns = ('-', '+', 'x', '\\', '*', 'o', 'O', '.')

VP2 = ax.boxplot(df_100, positions=[0.8], widths=0.25, patch_artist=True, showmeans=True, showfliers=True, medianprops={"color": "c", "linewidth": 1.0},
                 boxprops={"facecolor": "white", "edgecolor": "black", "linewidth": 0.5}, whiskerprops={"color": "black", "linewidth": 1.5}, capprops={"color": "black", "linewidth": 1.5})
for box in VP2['boxes']:
    box.set(hatch = 'o')

VP3 = ax.boxplot(df_300, positions=[1.1], widths=0.25, patch_artist=True, showmeans=True, showfliers=True, medianprops={"color": "c", "linewidth": 1.0}, 
                 boxprops={"facecolor": "white", "edgecolor": "black", "linewidth": 0.5}, whiskerprops={"color": "black", "linewidth": 1.5}, capprops={"color": "black", "linewidth": 1.5})
for box in VP3['boxes']:
    box.set(hatch = '+')

VP4 = ax.boxplot(df_500, positions=[1.4], widths=0.25, patch_artist=True, showmeans=True, showfliers=True, medianprops={"color": "c", "linewidth": 1.0},
                 boxprops={"facecolor": "white", "edgecolor": "black", "linewidth": 0.5}, whiskerprops={"color": "black", "linewidth": 1.5}, capprops={"color": "black", "linewidth": 1.5})
for box in VP4['boxes']:
    box.set(hatch = '/')

VP5 = ax.boxplot(df_700, positions=[1.7], widths=0.25, patch_artist=True, showmeans=True, showfliers=True, medianprops={"color": "c", "linewidth": 1.0}, 
                 boxprops={"facecolor": "white", "edgecolor": "black", "linewidth": 0.5}, whiskerprops={"color": "black", "linewidth": 1.5}, capprops={"color": "black", "linewidth": 1.5})
for box in VP5['boxes']:
    box.set(hatch = '-')

VP6 = ax.boxplot(df_900, positions=[2.0], widths=0.25, patch_artist=True, showmeans=True, showfliers=True, medianprops={"color": "c", "linewidth": 1.0},
                 boxprops={"facecolor": "white", "edgecolor": "black", "linewidth": 0.5}, whiskerprops={"color": "black", "linewidth": 1.5}, capprops={"color": "black", "linewidth": 1.5})
for box in VP6['boxes']:
    box.set(hatch = 'x')

ax.set(ylim=(-0.05, 1.05))

plt.xticks([0.8, 1.1, 1.4, 1.7, 2.0], ['100m', '300m', '500m', '700m', '900m'], fontsize = FS-4)
plt.ylabel('SSIM of Geolife', fontsize=FS)
plt.yticks(fontsize = FS)

patch_1 = mpatches.Patch(edgecolor = 'black', facecolor = 'white', hatch = 'o', label='100m')
patch_2 = mpatches.Patch(edgecolor = 'black', facecolor = 'white', hatch = '+', label='300m')
patch_3 = mpatches.Patch(edgecolor = 'black', facecolor = 'white', hatch = '/', label='500m')
patch_4 = mpatches.Patch(edgecolor = 'black', facecolor = 'white', hatch = '-', label='700m')
patch_5 = mpatches.Patch(edgecolor = 'black', facecolor = 'white', hatch = 'x', label='900m')

plt.show()
```
